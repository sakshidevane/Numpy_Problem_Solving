```python
import numpy as np
# sudoko  , rows and column sum should be 45 as 1 to 9 gives 45 , also 3*3 matrix total sum should be 45 then this called valid sudoku
s = np.array([[5,3,4,6,7,8,9,1,2],
             [6,7,2,1,9,5,3,4,8],
             [1,9,8,3,4,2,5,6,7],
             
             [8,5,9,7,6,1,4,2,3],
             [4,2,6,8,5,3,7,9,1],
             [7,1,3,9,2,4,8,5,6],
             
             [9,6,1,5,3,7,2,8,4],
             [2,8,7,4,1,9,6,3,5],
             [3,4,5,2,8,6,1,7,9]
            ])
```

```python

row_sum = np.sum(s, axis=1)
row_sum
```

```python
for i in row_sum:
    if i != 45:
        print(" not valid rows")
        break
else:
    print(" valid for rows ")
```

```python
col_sum = np.sum(s, axis=0)
col_sum
```

```python
result= (col_sum == col_sum[0]).all()
#another method insteda of for loop to check all element are same
```

```python
result
```

```python
if result ==True:
    print ("valid for column sudoku")
else :
    print("not valid for column")
```

```python
for i in range(0,9,3):
    for j in range(0,9,3):
        small_matrix = s[i:i+3,j:j+3]
        # print(small_matrix)
        n =np.sum(small_matrix)
        # print(n)
        if n!=45:
            print ("invalid sudoku")
            break
else:
    print ("valid sudoku")
    
```
